### Basic Steps of the Data Analyst

**1] Finding the Null Values in thee dataset**

**2] Explore about the numerical feature**

**3] Explore about categorical feature**

**4] Findinging relationshiop between feature** 

To check any null value present in the dataset or not

```python
dataset.isnull().sum()
```

Find missing value column with the help of code

```python
[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
```

To Display the missing values of features with graph

```python
msno.bar(dataset)
```

We have 9 missing values in the Cuisines

Matrix Display of Missing value

```python
msno.matrix(dataset)
```

Display the missing values by using the heatmap

```python
plt.figure(figsize= (8,8))
sns.heatmap(dataset.isnull(),cmap='viridis')
```

Check any Duplicate Record is present in the dataset or not 

```python
dataset[dataset.duplicated()]
```

To check total no of duplicate present in the dataset

```python
dataset.duplicated().sum()
```
